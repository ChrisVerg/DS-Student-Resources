# US Government Data Catalogs

[Catalog.data.gov](https://catalog.data.gov/dataset): This is the official data catalog for the U.S. government, and it provides access to over 250,000 datasets from various federal agencies.

[USASpending.gov](https://www.usaspending.gov/): This website provides access to data on federal spending, including contracts, grants, and loans.

[HealthData.gov](https://healthdata.gov/): This website provides access to health-related data from various federal agencies, including the Centers for Disease Control and Prevention (CDC) and the National Institutes of Health (NIH).

[Climate.gov](https://www.climate.gov/): This website provides access to data related to climate, including maps, charts, and datasets.

[Census.gov](https://www.census.gov/): This website provides access to data from the U.S. Census Bureau, including population, housing, and economic data.

[Energy.gov](https://www.energy.gov/data/open-energy-data): This website provides access to data related to energy, including renewable energy, energy efficiency, and fossil fuels.

[National Archives Catalog](https://catalog.archives.gov/): This website provides access to a wide range of historical and government records, including photographs, documents, and audio and video recordings.

[NASA Open Data Portal](https://data.nasa.gov/): This website provides access to data related to space exploration and science, including satellite imagery, climate data, and astronaut biographies.

[National Oceanic and Atmospheric Administration (NOAA) Data Catalog](https://data.noaa.gov/datasetsearch/): This website provides access to a wide range of environmental data, including weather forecasts, oceanographic data, and climate models.

[Federal Reserve Economic Data (FRED)](https://fred.stlouisfed.org/): This website provides access to economic data from the Federal Reserve System, including data on employment, inflation, and interest rates.

[USAFacts.org](https://usafacts.org/): This website provides access to data related to government finance and spending, including data on taxes, revenue, and government programs.

[Department of Education's Ed Data Express](https://eddataexpress.ed.gov/): This website provides access to data related to education, including data on graduation rates, enrollment, and student achievement.

[National Cancer Institute's Cancer Data Access System (CDAS)](https://prevention.cancer.gov/resources/cancer-data-access-system): This website provides access to data related to cancer research, including clinical trial data and cancer incidence and mortality rates.

[Department of Labor's Data & Statistics](https://www.bls.gov/): This website provides access to data related to labor and employment, including data on wages, occupational safety, and workforce demographics.

[National Renewable Energy Laboratory's (NREL) Open Data](https://www.nrel.gov/research/data-tools.html): This website provides access to data related to renewable energy, including data on solar, wind, and geothermal energy resources.

[Department of Transportation's National Transportation Library](https://ntl.bts.gov/ntl): This website provides access to data related to transportation, including data on highways, airports, and public transportation.

[National Institute of Standards and Technology (NIST) Data Gateway](https://www.nist.gov/publications/nist-data-gateway): This website provides access to data related to science and technology, including data on materials properties, energy efficiency, and atomic spectra.

[National Library of Medicine's (NLM) National Center for Biotechnology Information (NCBI)](https://www.ncbi.nlm.nih.gov/): This website provides access to data related to biomedical research, including data on genetics, genomics, and clinical trials.

[Department of Homeland Security's (DHS) Data & Statistics](https://www.dhs.gov/data): This website provides access to data related to homeland security, including data on immigration, border security, and disaster response.

[United States Geological Survey's (USGS) Science Data Catalog](https://data.usgs.gov/datacatalog/): This website provides access to data related to natural resources, hazards, and environmental health, including data on earthquakes, volcanoes, and water resources.

[Environmental Protection Agency's (EPA) Environmental Data Gateway](https://edg.epa.gov/metadata/catalog/main/home.page): This website provides access to data related to environmental quality, including data on air pollution, water quality, and toxic substances.

[Department of Agriculture's (USDA) National Agricultural Library](https://www.nal.usda.gov/): This website provides access to data related to agriculture, including data on crop yields, food safety, and rural development.

[Centers for Medicare & Medicaid Services (CMS) Data Navigator](https://data.cms.gov/): This website provides access to data related to healthcare, including data on hospital quality, healthcare spending, and health insurance coverage.

[Department of Housing and Urban Development's (HUD) Open Data Portal](https://data.hud.gov/): This website provides access to data related to housing and urban development, including data on affordable housing, community development, and homelessness.

[Department of Veterans Affairs' (VA) Open Data Portal](https://www.data.va.gov/): This website provides access to data related to veterans' health, benefits, and services, including data on veteran population demographics, healthcare utilization, and disability compensation.

[Consumer Financial Protection Bureau's (CFPB) Open Data](https://www.consumerfinance.gov/data-research/): This website provides access to data related to consumer finance, including data on mortgage lending, credit card usage, and payday lending.

[Department of Commerce's International Trade Administration (ITA) Data and Analysis](https://www.trade.gov/trade-data-analysis): This website provides access to data related to international trade and investment, including data on export and import activity, trade agreements, and foreign direct investment.

[National Agricultural Statistics Service's (NASS) Quick Stats](https://www.nass.usda.gov/Quick_Stats/): This website provides access to data related to agriculture and rural communities, including data on crop production, livestock inventory, and farm income.

[Department of Energy's (DOE) Energy Information Administration (EIA) Open Data Portal](https://www.eia.gov/opendata/): This website provides access to data related to energy markets and use, including data on energy production and consumption, electricity generation, and petroleum prices.

[National Science Foundation's (NSF) National Center for Science and Engineering Statistics (NCSES)](https://ncses.nsf.gov/explore-data): This website provides access to data related to science and engineering research and development, including data on research funding, workforce trends, and scientific publications.

[National Park Service's (NPS) Data Store](https://irma.nps.gov/DataStore/): This website provides access to data related to the natural and cultural resources of the National Park System, including data on park visitation, wildlife populations, and cultural artifacts.

[Department of Justice's (DOJ) Bureau of Justice Statistics (BJS)](https://bjs.ojp.gov/): This website provides access to data related to crime and justice, including data on law enforcement personnel, crime rates, and the criminal justice system.

[National Highway Traffic Safety Administration's (NHTSA) Traffic Safety Fact Sheets](https://crashstats.nhtsa.dot.gov/#!/DocumentTypeList/11): This website provides access to data related to traffic safety, including data on motor vehicle crashes, fatalities, and injuries.

[Department of State's Office of the Historian](https://history.state.gov/): This website provides access to data related to US diplomatic history, including data on foreign relations, international treaties, and diplomatic correspondence.

[General Services Administration's (GSA) Data.gov Catalog](https://catalog.data.gov/organization/gsa-gov): This website provides access to a wide range of government data, including data on federal spending, health, education, and more.

[Department of Defense's (DoD) Defense Technical Information Center (DTIC)](https://discover.dtic.mil/): This website provides access to data related to defense research and development, including data on weapons systems, military technology, and scientific research.

[Department of Homeland Security's (DHS) Homeland Security Digital Library (HSDL)](https://www.hsdl.org/c/): This website provides access to data related to homeland security, including data on emergency management, cybersecurity, and border security.

[National Aeronautics and Space Administration's (NASA) Earth Observing System Data and Information System (EOSDIS)](https://www.earthdata.nasa.gov/eosdis): This website provides access to data related to Earth science research, including data on climate change, atmospheric chemistry, and land use.

[National Archives and Records Administration's (NARA) Archives.gov](https://www.archives.gov/): This website provides access to historical records and government documents, including data on military records, genealogy, and presidential libraries.

[Department of the Interior's (DOI) Bureau of Land Management (BLM) Data Services](https://www.blm.gov/services/geospatial/GISData): This website provides access to data related to public lands and natural resources, including data on land ownership, minerals, and wildlife.

[Department of Transportation's (DOT) Bureau of Transportation Statistics (BTS)](https://www.bts.gov/): This website provides access to data related to transportation, including data on air travel, freight transportation, and traffic congestion.
