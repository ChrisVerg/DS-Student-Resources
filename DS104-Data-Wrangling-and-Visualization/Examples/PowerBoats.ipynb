```r
library(readr)
library(ggplot2)
library(lattice)
```

```r
powerboats= read_csv("/Users/julieraeboucher/Downloads/L3P1.csv")
```

```r
head(powerboats)
```

```r
# Basic Histogram:
ggplot(powerboats, aes(`Power Boats (in 1000's)`)) + geom_histogram(binwidth= 30, fill="deepskyblue4", color="white")
```

```r
# Labeled Histogram:
ggplot(powerboats, aes(`Power Boats (in 1000's)`)) +
geom_histogram() +
ggtitle("Title") +
xlab("X Label") +
ylab("Y Label / Frequency")
```

```r
# Labeled Bar:
ggplot(powerboats, stat_count(data = powerboats$`Power Boats (in 1000's)`, mapping = NULL, stat = "identity")) +
geom_bar() +
ggtitle("Title") +
xlab("X Label") +
ylab("Y Label / Frequency")
```

```r
# library(lattice)
barchart(powerboats$`Power Boats (in 1000's)`, main= "Title", ylab= "Y Label" , xlab= "X Label / Frequency")
# (By default x and y are switched in lattice
```
