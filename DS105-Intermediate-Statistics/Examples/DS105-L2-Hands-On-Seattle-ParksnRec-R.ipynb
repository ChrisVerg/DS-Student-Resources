#### increase/decrease of prior year

```r
plotNormalHistogram(data$increase.decrease.of.prior.year)
## looks normally distributed
```

```r
data$increase.decrease.of.prior.yearSQRT = sqrt(data$increase.decrease.of.prior.year)
## taking square root like it is positively skewed
```

```r
data$increase.decrease.of.prior.yearSQ = data$increase.decrease.of.prior.year ^ 2
## squaring it like it is negatively skewed
```

```r
plotNormalHistogram(data$increase.decrease.of.prior.yearSQRT)
## maybe more positively skewed than before - original is best
```

```r
plotNormalHistogram(data$increase.decrease.of.prior.yearSQ)
## definitely made positively skewed - origincal is best
```
